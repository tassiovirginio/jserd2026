```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Carregar os dados APENAS do CSV
df = pd.read_csv("testsmellsxsentiments.csv")

# 2. Ordenao: Garante que o valor "mais negativo" fique  esquerda
df = df.sort_values(by='total_score', ascending=True)

# Criando o grfico
plt.figure(figsize=(20, 10))

# Plotando barras verticais (valores negativos para baixo)
bars = plt.bar(df['testsmell'], df['total_score'], color='skyblue', edgecolor='black', alpha=0.8)

# Ajustes de Eixos e Ttulos
plt.axhline(0, color='black', linewidth=1) # Linha de base (eixo X)
plt.xticks(rotation=45, ha="right", fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel("Sentiment Total Score", fontsize=16, labelpad=15)
plt.title("Impact of Test Smells on Sentiment Score (Sum)", fontsize=20, pad=20)

# Limpeza visual (remover bordas desnecessrias)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

# Grid apenas no eixo Y
plt.grid(axis='y', linestyle='--', alpha=0.5)

# 3. RTULOS FORMATADOS
offset = df['total_score'].min() * 0.02 # Pequeno espao calculado dinamicamente

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2, 
        height - abs(offset), # Posio abaixo da barra
        f"{int(height):,}".replace(",", "."), # Formatao PT-BR (ponto no milhar)
        ha='center', 
        va='top', 
        fontsize=12,
        fontweight='bold',
        color='black'
    )

# Ajuste de layout para evitar cortes nos nomes
plt.subplots_adjust(bottom=0.25)

# Salvar
plt.savefig("testsmells_sentiment_chart.png", format="png", bbox_inches="tight", dpi=300)

plt.show()
```
