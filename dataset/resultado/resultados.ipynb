```python
%load_ext sql
%sql sqlite:///resultado.sqlite

import pandas as pd
import sqlite3
print(sqlite3.sqlite_version)

conn = sqlite3.connect("resultado.sqlite")  # ou :memory:
pd.read_sql_query("""
SELECT name AS table_name
FROM sqlite_master
WHERE type = 'table'
ORDER BY name
""", conn)
```

```python
pd.read_sql_query("""

-- Score de negatividade por testsmells
SELECT distinct (testsmell),
    sum(score)
FROM testsmells
where
    score < 0
group by
    testsmell;

""", conn)
```

```python
pd.read_sql_query("""
-- média dos sentimentos negativos por testsmell

SELECT testsmell, AVG(CAST(score AS REAL)) as media_score
FROM testsmells
GROUP BY
    testsmell
ORDER BY media_score ASC;

""", conn)
```

```python
pd.read_sql_query("""


SELECT testsmell, COUNT(score) as quantidade
FROM testsmells
WHERE
    CAST(score AS REAL) < 0
GROUP BY
    testsmell;


""", conn)
```

```python
pd.read_sql_query("""

-- soma dos sentimentos negativos
SELECT testsmell, SUM(CAST(score AS REAL)) as total_score
FROM testsmells
WHERE
    CAST(score AS REAL) < 0
GROUP BY
    testsmell
ORDER BY total_score ASC;

""", conn)
```

```python
pd.read_sql_query("""

-- média dos sentimentos negativos por testsmell
SELECT testsmell, SUM(CAST(score AS REAL)) / COUNT(score) as media
FROM testsmells
WHERE
    CAST(score AS REAL) < 0
GROUP BY
    testsmell
ORDER BY media ASC;

""", conn)
```

```python
pd.read_sql_query("""

-- soma dos sentimentos negativos por autor
SELECT author, SUM(CAST(score AS REAL)) as total_score
FROM testsmells
WHERE
    CAST(score AS REAL) < 0
GROUP BY
    author
ORDER BY total_score ASC;

""", conn)
```

```python
pd.read_sql_query("""

-- média dos sentimentos negativos por autor
SELECT author, SUM(CAST(score AS REAL)) / COUNT(score) as media
FROM testsmells
WHERE
    CAST(score AS REAL) < 0
GROUP BY
    author
ORDER BY media ASC;

""", conn)
```

```python
pd.read_sql_query("""

-- quantidade de projetos por autor
SELECT author, COUNT(DISTINCT project_name) as quantidade
FROM testsmells
GROUP BY
    author
ORDER BY quantidade DESC limit 20;

""", conn)
```
