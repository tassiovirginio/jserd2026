```python
import numpy as np
import matplotlib.pyplot as plt


# Dados dos test smells
test_smells = {
    "Assertion Roulette": 246247,
    "Conditional Test Logic": 11605,
    "Duplicate Assert": 378527,
    "Empty Test": 471,
    "Exception Handling": 4013,
    "Ignored Test": 285,
    "Magic Number": 427935,
    "Print Statment Fixture": 3795,
    "Resource Optimism": 3640,
    "Sensitive Equality": 7526,
    "Sleepy Fixture": 2313,
    "Test Without Description": 240,
    "Unknown Test": 16927,
    "Verbose Test": 12414,
}

# Criando o grfico
plt.figure(figsize=(20, 8))  # Aumentei a altura para dar mais espao
plt.bar(test_smells.keys(), test_smells.values(), color='skyblue')

# Ajustando tamanho da fonte das labels
plt.xticks(rotation=45, ha="right", fontsize=14)  # Fonte maior para o eixo X
plt.yticks(fontsize=14)  # Fonte maior para o eixo Y
plt.ylabel("Sum of Test Smells", fontsize=16)  # Fonte maior para o rtulo do eixo Y
plt.title("Distribution of Test Smells", fontsize=18)  # Fonte maior para o ttulo

# Exibir os valores no topo das barras
for i, v in enumerate(test_smells.values()):
    plt.text(i, v + 0.1, f"{v:.0f}", ha='center', fontsize=12)  # Fonte maior nos valores

# Ajustar layout para evitar cortes
plt.subplots_adjust(bottom=0.3)  # Garante espao para os rtulos no eixo X

# Salvar a imagem sem cortes
plt.savefig("distribuicao_linear.png", format="png", bbox_inches="tight", pad_inches=0.2)

plt.show()

```

```python

# Aplicando log na base 10
log_values = {key: np.log10(value) for key, value in test_smells.items()}
log_values

```

```python
import matplotlib.pyplot as plt

# Criando o grfico
plt.figure(figsize=(20, 9))  # Aumentei a altura para dar mais espao
plt.bar(log_values.keys(), log_values.values(), color='skyblue')
plt.xticks(rotation=45, ha="right")
plt.xticks(rotation=45, ha="right", fontsize=14)  # Fonte maior para o eixo X
plt.yticks(fontsize=14)  # Fonte maior para o eixo Y
plt.ylabel("log10(Sum of Test Smells)")
plt.title("Distribution of Test Smells (Logarithmic Scale))")

# Exibir os valores no topo das barras
for i, v in enumerate(log_values.values()):
    plt.text(i, v + 0.1, f"{v:.2f}", ha='center', fontsize=10)

# Ajustar layout para evitar cortes
plt.subplots_adjust(bottom=0.3)  # Garante espao para os rtulos no eixo X

# Salvar a imagem sem cortes
plt.savefig("distribuicao_log10.png", format="png", bbox_inches="tight", pad_inches=0.2)

plt.show()
```
